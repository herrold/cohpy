This notebook explores the order of items in sets.
It seems that when Jupyter prints out the value of a set,
if the set has less than 1000 items
the output is sorted.
Sets with 1000 or more items seem to not be sorted.

This seems to be just Jupyter behavior, not of Python itself.
str(b), repr(b), print(b), print(str(b)), and print(repr(b))
are seem to always be unordered.
